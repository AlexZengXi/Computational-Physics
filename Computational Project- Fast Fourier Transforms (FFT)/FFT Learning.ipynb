```python
# the Python implementation of FFT
# Alex Zeng, 1007099373, Jan 20th 2023

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

# Generating 38 data points of a time series of sin(2.5 t) with \Delta = 0.20
tmin = 0
tmax = 2.4 * sp.pi
delta = 0.2
t = np.arange(tmin, tmax, delta)
y = np.sin(2.5 * t)

plt.figure(6)
plt.plot(t, y, 'bo')
plt.title('y(t) = sin(2.5 t)')
plt.xlabel('t (s)')
plt.ylabel('y(t)')
plt.show()
```

```python
from scipy.fft import fft

# compute the discrete Fourier Transform using the FFT routine
Y = fft(y)

plt.figure(7)
plt.plot(np.imag(Y))
plt.title('Imaginary part of the Fourier transform of sin(2.5 t)')
plt.xlabel('j')
plt.ylabel("Yj")
plt.show()
```

```python
from matplotlib.pyplot import figure, plot, title, xlabel, ylabel, show
from numpy import pi, imag
from numpy.fft import fftfreq

n = len(y)
# Calculate frequencies of the transform in Hz
freq = fftfreq(n, delta)
# Convert to angular frequencies, since the entire document is using it
w = 2 * pi * freq

figure(8)
plot(w, imag(Y))
title('Imaginary part of the Fourier transform of sin(2.5 t)')
xlabel('w (rad/s)')
ylabel('Yj')
show()

# Now we can see the positive and negative peak from the graph generated
```
